```python
import pandas as pd
import numpy as np
import seaborn as sn

data = pd.read_csv ('C:/Users/joelk/IRONHACK_CLASS/World-energy-production-and-consumption-analysis/Data/cleaned_energy_data.csv')
data
```

```python
import matplotlib.pyplot as plt

# Filter the data for the years 2000 and 2020
df_filtered = data[data['Year'].isin([2000, 2020])]

# Select relevant columns (Country, Year, CO2 emissions)
df_filtered_co2 = df_filtered[['Country', 'Year', 'CO2_Emissions_MtCO2']]

# Sort the data by CO2 emissions in descending order and take the top 10 for each year
df_2000 = df_filtered_co2[df_filtered_co2['Year'] == 2000].sort_values(by='CO2_Emissions_MtCO2', ascending=False).head(10)
df_2020 = df_filtered_co2[df_filtered_co2['Year'] == 2020].sort_values(by='CO2_Emissions_MtCO2', ascending=False).head(10)

# Plot for the year 2000
plt.figure(figsize=(10, 6))
plt.barh(df_2000['Country'], df_2000['CO2_Emissions_MtCO2'], color='skyblue')
plt.xlabel('CO2 Emissions (MtCO2)')
plt.ylabel('Country')
plt.title('Top 10 Countries by CO2 Emissions in 2000')
plt.tight_layout()
plt.show()

# Plot for the year 2020
plt.figure(figsize=(10, 6))
plt.barh(df_2020['Country'], df_2020['CO2_Emissions_MtCO2'], color='lightgreen')
plt.xlabel('CO2 Emissions (MtCO2)')
plt.ylabel('Country')
plt.title('Top 10 Countries by CO2 Emissions in 2020')
plt.tight_layout()
plt.show()
```

```python
# Filter the data for the years 2000 and 2020
df_filtered = data[data['Year'].isin([2000, 2020])]

# Select relevant columns for energy production and consumption (Country, Year, Total Energy Production and Consumption)
df_filtered_energy = df_filtered[['Country', 'Year', 'Total_Energy_Production_Mtoe', 'Total_Energy_Consumption_Mtoe']]

# Sort the data by total energy production in descending order and take the top 10 for each year
df_prod_2000 = df_filtered_energy[df_filtered_energy['Year'] == 2000].sort_values(by='Total_Energy_Production_Mtoe', ascending=False).head(10)
df_prod_2020 = df_filtered_energy[df_filtered_energy['Year'] == 2020].sort_values(by='Total_Energy_Production_Mtoe', ascending=False).head(10)

# Sort the data by total energy consumption in descending order and take the top 10 for each year
df_cons_2000 = df_filtered_energy[df_filtered_energy['Year'] == 2000].sort_values(by='Total_Energy_Consumption_Mtoe', ascending=False).head(10)
df_cons_2020 = df_filtered_energy[df_filtered_energy['Year'] == 2020].sort_values(by='Total_Energy_Consumption_Mtoe', ascending=False).head(10)

# Plot for energy production in 2000
plt.figure(figsize=(10, 6))
plt.barh(df_prod_2000['Country'], df_prod_2000['Total_Energy_Production_Mtoe'], color='lightblue')
plt.xlabel('Energy Production (Mtoe)')
plt.ylabel('Country')
plt.title('Top 10 Countries by Energy Production in 2000')
plt.tight_layout()
plt.show()

# Plot for energy production in 2020
plt.figure(figsize=(10, 6))
plt.barh(df_prod_2020['Country'], df_prod_2020['Total_Energy_Production_Mtoe'], color='lightgreen')
plt.xlabel('Energy Production (Mtoe)')
plt.ylabel('Country')
plt.title('Top 10 Countries by Energy Production in 2020')
plt.tight_layout()
plt.show()

# Plot for energy consumption in 2000
plt.figure(figsize=(10, 6))
plt.barh(df_cons_2000['Country'], df_cons_2000['Total_Energy_Consumption_Mtoe'], color='salmon')
plt.xlabel('Energy Consumption (Mtoe)')
plt.ylabel('Country')
plt.title('Top 10 Countries by Energy Consumption in 2000')
plt.tight_layout()
plt.show()

# Plot for energy consumption in 2020
plt.figure(figsize=(10, 6))
plt.barh(df_cons_2020['Country'], df_cons_2020['Total_Energy_Consumption_Mtoe'], color='orange')
plt.xlabel('Energy Consumption (Mtoe)')
plt.ylabel('Country')
plt.title('Top 10 Countries by Energy Consumption in 2020')
plt.tight_layout()
plt.show()
```

```python

import seaborn as sns
# Filter the data for the year 2020
df_2020 = data[data['Year'] == 2020]

# Select relevant columns for energy production, energy consumption, and CO2 emissions
df_2020_corr = df_2020[['Total_Energy_Production_Mtoe', 'Total_Energy_Consumption_Mtoe', 'CO2_Emissions_MtCO2']]

# Calculate the correlation matrix
correlation_matrix = df_2020_corr.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap for Energy Consumption, Production, and CO2 Emissions (2020)')
plt.show()
```

```python
# Filter the data for Europe and for the years 2000 to 2020
df_europe = data[(data['Region'] == 'Europe') & (data['Year'].between(2000, 2020))]

# Pivot the data so that each country becomes a column and the rows represent years
df_europe_pivot = df_europe.pivot_table(index='Year', columns='Country', values='Total_Energy_Consumption_Mtoe')

# Plot the energy consumption trend for each European country
plt.figure(figsize=(12, 8))

# Plot each country's energy consumption as a separate line
for country in df_europe_pivot.columns:
    plt.plot(df_europe_pivot.index, df_europe_pivot[country], marker='', linestyle='-', label=country)

plt.xlabel('Year')
plt.ylabel('Total Energy Consumption (Mtoe)')
plt.title('Energy Consumption Trend by European Countries (2000-2020)')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Filter the data for Europe and for the year 2020
df_europe_2020 = data[(data['Region'] == 'Europe') & (data['Year'] == 2020)]

# Sort the data by energy consumption for better visualization
df_europe_2020_sorted = df_europe_2020.sort_values(by='Total_Energy_Consumption_Mtoe', ascending=False)

# Define the positions and width for the bars
x = np.arange(len(df_europe_2020_sorted['Country']))  # the label locations
width = 0.35  # the width of the bars

# Create the bar chart for both consumption and production
fig, ax = plt.subplots(figsize=(12, 8))

# Bar chart for energy consumption
bar1 = ax.bar(x - width/2, df_europe_2020_sorted['Total_Energy_Consumption_Mtoe'], width, label='Consumption', color='skyblue')

# Bar chart for energy production
bar2 = ax.bar(x + width/2, df_europe_2020_sorted['Total_Energy_Production_Mtoe'], width, label='Production', color='lightgreen')

# Add labels, title, and custom x-axis tick labels
ax.set_xlabel('Country')
ax.set_ylabel('Energy (Mtoe)')
ax.set_title('Comparison of Energy Consumption and Production by European Countries in 2020')
ax.set_xticks(x)
ax.set_xticklabels(df_europe_2020_sorted['Country'], rotation=90)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
